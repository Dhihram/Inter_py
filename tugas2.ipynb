```python
index = ['Start Date', 'End Date', "Category", "Start City", "Stop City", "Miles", "Purpose"]
df.columns = index
df
```

Kategorisasi Cara I
menggunakan numpy

```python
 conditions = [
    ((df['Miles'] < 5)),
    ((df['Miles'] >= 5)& (df['Miles']<=10)),
    ((df['Miles'] > 10))
]
choices = ['Short trip','Medium trip','Long trip']
```

```python
df['Miles_cat'] = np.select(conditions, choices, default='NA')
df
```

Kategorisasi Cara II Menggunakan Loc

```python
def kat(df):
    df.loc[(df['Miles'] < 5),'Miles_cat'] = 'Short trip'
    df.loc[(df['Miles'] >= 5) & (df['Miles'] <= 10) ,'Miles_cat'] = 'Medium trip'
    df.loc[(df['Miles'] > 10),'Miles_cat'] = 'Long trip'
    df['Miles_cat'].fillna('C', inplace=True)
```

```python
kat(df)
df
```

Kategorisasi Cara III Menggunakan If

```python
def kat_fi(df):    
    if (df['Miles'] < 5):
        return 'Short trip'
    elif (df['Miles'] >= 5) & (df['Miles'] <= 10):
        return 'Medium trip'
    elif (df['Miles'] > 10):
        return "Long trip"	
```

```python
df['miles_cat'] = df.apply(kat_fi, axis = 1)
```
